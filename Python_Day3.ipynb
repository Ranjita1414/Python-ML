# **Numpy**

**Create Arrays**

```python
import numpy as np

a=np.array([3,6,2,8,4])
b=np.arange(2,4)
c=np.zeros((1,4))
d=np.ones((4,5))

print("Array a:",a)
print("Array b:",b)
print("1x4 Zeros:",c)
print("4x5 Ones:",d)
```

- Indexing And Slicing

```python
print("First Element of a:",a[0])
print("Last Element of a:",a[-2:])
```

- Basic Arithmetic Operations

```python
print("a+2:",a+2)
print("a*3:",a*3)
```

- Basic statistics

```python
print("Square of a:",np.square(a))
print("Mean of a:",np.mean(a))
print("Standard divition:",np.std(a))
```

- Reshaping arrays

```python
matrix=np.arange(1,10).reshape((3,3))
print("Reshaping 3x3 matrix:")
matrix
```

- **Transpose**

```python
print("Transpose")
matrix.T
```

- Boolean indexing

```python
print("Elements>=5:",matrix[matrix>=5])
```

# **Pandas**

**Create a DataFrame manually**

```python
import pandas as pd

data={
    'Name':['Alice','Bob','Charlie','devid','Eve'],
    'Age':[25,30,40,22,29],
    'Department':['Sales','Engineering','HR','Engineering','Sales'],
    'Salary':[50000,80000,40000,90000,45000]
}

df=pd.DataFrame(data)
df
```

**Basic operations**

```python
print("First 3 Rows:")
df.head(3)
```

```python
print("Coloumn name:")
df.columns
```

```python
print("Data Types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

- Filtering

```python
print("People with salary >=50000:")
df[df['Salary']>=50000]
```

- Sorting

```python
print("Sorted by Age:")
df.sort_values(by="Age")
```

- **Groupping and aggregation**

```python
grouped=df.groupby('Department')['Salary'].mean()
print("Average salary by Department:")
grouped
```

- **Adding new columns**

```python
df['Salary in Lakhs']=df['Salary']/100000
df
```

# **MATPLOT**

- **LinePlot**

```python
import matplotlib.pyplot as plt

years=[2019,2020,2021,2022]
revenue=[100,120,130,160]

plt.plot(years,revenue,marker='<',linestyle='--',color='purple')
plt.title("Company Revenue over years")
plt.xlabel("Year")
plt.ylabel("Revenue(in crores)")
plt.xticks(years)
plt.grid(False)
plt.show()
```

**Bar Chart**

```python
departments=df['Department'].value_counts()
departments.plot(kind='bar',color='teal')
plt.title("Department Distribution")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

**Histogram**

```python
ages=df['Age']
plt.hist(ages,bins=5,color='orange',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

**Piechart**

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```

**SEABORN**

- **Load sample dataset**

```python
import seaborn as sns

tips=sns.load_dataset("tips")
tips.head()
```

```python
print("\nInfo about the tips dataset:")
tips.info()

print("\nDescription of the tips dataset:")
tips.describe(include='all')
```

**Pairplot**

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

- **HEATMAP**

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.countplot(x='day',data=tips,palette='Set2',hue='day',legend=True)
plt.title("Counts of Records by day")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Boxplot of Total Bill by day")
plt.show()
```
